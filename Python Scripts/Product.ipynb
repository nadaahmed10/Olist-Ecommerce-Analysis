```python
import pandas as pd
import numpy as np 


dataset['product_category_name'] = dataset['product_category_name'].str.title()

####### Create Category For all dataset ########
conditions = [
(dataset['product_category_name'].isin(['Fashion_Calcados',
                                    'Fashion_Bolsas_E_Acessorios',
                                    'Fashion_Roupa_Masculina',
                                    'Fashion_Roupa_Feminina',
                                    'Fashion_Roupa_Infanto_Juvenil',
                                    'Fashion_Underwear_E_Moda_Praia',
                                    'Fashion_Esporte'])),

(dataset['product_category_name'].isin(['Cama_Mesa_Banho',
                                    'Moveis_Decoracao',
                                    'Moveis_Sala',
                                    'Moveis_Cozinha_Area_De_Servico_Jantar_E_Jardim',
                                    'Moveis_Quarto',
                                    'Moveis_Escritorio',
                                    'Moveis_Colchao_E_Estofado',
                                    'Casa_Construcao',
                                    'Casa_Conforto',
                                    'Casa_Conforto_2'])),

(dataset['product_category_name'].isin(['Informatica_Acessorios',
                                    'Eletronicos'
                                    ,'Pc_Gamer'
                                    ,'Telefonia'
                                    ,'Telefonia_Fixa'
                                    ,'Tablets_Impressao_Imagem'
                                    ,'Audio'
                                    ,'Cine_Foto'
                                    ,'Consoles_Games'
                                    ,'Eletrodomesticos'
                                    ,'Eletrodomesticos_2'
                                    ,'Eletroportateis'
                                    ,'Portateis_Casa_Forno_E_Cafe'
                                    ,'Portateis_Cozinha_E_Preparadores_De_Alimentos'])),

(dataset['product_category_name'].isin(['Brinquedos','Bebes','Fraldas_Higiene'])) ,

(dataset['product_category_name'].isin(['Beleza_Saude','Perfumaria'])) ,

(dataset['product_category_name'].isin(['Pet_Shop'])) ,

(dataset['product_category_name'].isin(['Esporte_Lazer'])) ,

(dataset['product_category_name'].isin(['Ferramentas_Jardim',
                                    'Construcao_Ferramentas_Construcao',
                                    'Construcao_Ferramentas_Seguranca',
                                    'Construcao_Ferramentas_Iluminacao',
                                    'Construcao_Ferramentas_Ferramentas',
                                    'Construcao_Ferramentas_Jardim',
                                    'Industria_Comercio_E_Negocios'])) ,

(dataset['product_category_name'].isin(['Livros_Interesse_Geral'
                                    ,'Livros_Importados'
                                    ,'Livros_Tecnicos'
                                    ,'Dvds_Blu_Ray'
                                    ,'Cds_Dvds_Musicais'
                                    ,'Musica'])) ,

(dataset['product_category_name'].isin(['Alimentos','Alimentos_Bebidas','Bebidas'])) ,

(dataset['product_category_name'].isin(['Relogios_Presentes'
                                    ,'Cool_Stuff'
                                    ,'Artigos_De_Festas'
                                    ,'Artigos_De_Natal'
                                    ,'Flores'
                                    ,'Malas_Acessorios'])) ,

(dataset['product_category_name'].isin(['Papelaria'])) ,

(dataset['product_category_name'].isin(['Seguros_E_Servicos' ,'Market_Place'])) ,

(dataset['product_category_name'].isin(['Artes' , 'Artes_E_Artesanato'])) , 
(dataset['product_category_name'].isin(['Automotivo'])) , 
(dataset['product_category_name'].isin(['Unknown']))
]

choices = [
'Fashion',
'Home & Furniture',
'Electronics & Tech' ,
'Kids & Baby' , 
'Beauty & Health' ,
'Pets' ,
'Sports' ,
'Construction Tools' ,
'Books & Media' ,
'Food & Drinks' ,
'Gifts' ,
'Stationery' ,
'Services' ,
'Arts' ,
'Automotive' ,
'Unknown'
]
dataset['Product_Category'] = np.select(conditions, choices, default='Other')

##### Correct Some dataset ######

category_corrections = {
# Typos
'fashio_female_clothing': 'fashion_female_clothing',
'costruction_tools_garden': 'construction_tools_garden',
'costruction_tools_tools': 'construction_tools_tools',
'home_confort': 'home_comfort',

# Merging duplicates/versions
'home_appliances_2': 'home_appliances',
'home_comfort_2': 'home_comfort'}

dataset['Product_Category'] = dataset['Product_Category'].replace(category_corrections)

##### Filling the Null Value for Kids and Unknown dimensions #####

cols = ['product_weight_g','product_length_cm','product_height_cm','product_width_cm']

dataset[cols] = dataset.groupby('Product_Category')[cols].transform(lambda x: x.fillna(x.median()))


##### Product Weight in kilos #####

dataset['product_weight_k'] = dataset['product_weight_g'] / 1000


##### Dropping unnecessary columns #####

dataset.drop(columns=['product_name_lenght','product_description_lenght','product_photos_qty','product_category_name','product_weight_g'],inplace=True)

##### Cleaning column names #####

dataset.columns = dataset.columns.str.replace('_', ' ').str.title()

output = dataset
```
